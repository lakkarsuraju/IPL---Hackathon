### 8. Which statement holds true?

```python
# Clculate the number of matches won by teams choosing to bat and field
bat_wins = ipl_df[ipl_df['toss_decision'] == 'bat']['winner'].count()
field_wins = ipl_df[ipl_df['toss_decision'] == 'field']['winner'].count()

# Compare the counts and determine which statement holds true
if bat_wins > field_wins:
    print("B) Teams winning tosses and electing bat first have won more number of matches")
elif field_wins > bat_wins:
    print("A) Teams winning tosses and electing field first have won more number of matches")
else:
    print("C) both A and B")
       
```

### 10. To plot the most consistent batsmen among top 10 run getters, what plotting technique we use?

```python
# Assuming match_df and del_df are already loaded

# Step 1: Calculate the total runs scored by each batsman
total_runs_by_batsman = del_df.groupby('batsman')['total_runs'].sum().reset_index()

# Step 2: Get the top 10 run-getters
top_10_run_getters = total_runs_by_batsman.nlargest(10, 'total_runs')

# Step 3: Filter the delivery DataFrame for the top 10 run-getters
del_df_top_10 = del_df[del_df['batsman'].isin(top_10_run_getters['batsman'])]

# Step 4: Plot the runs scored by each batsman over matches using a line plot
plt.figure(figsize=(12, 8))
for batsman in top_10_run_getters['batsman']:
    batsman_data = del_df_top_10[del_df_top_10['batsman'] == batsman]
    plt.plot(batsman_data.groupby('match_id')['total_runs'].cumsum(), label=batsman, marker='o')

plt.title('Runs Scored by Top 10 Batsmen Over Matches')
plt.xlabel('Matches')
plt.ylabel('Total Runs')
plt.legend()
plt.show()
```
